```python
import os
from netpyne import sim
from assert_static_simcfg import revise_cell_params
data_path = os.path.abspath('simulation_with_only_cand_of_interest/gen_1/gen_1_cand_29_data.json')
temp_netParams_path = 'temp_netParams.json'
extracted_data_path = 'extracted_data.json'
cfg_path = os.path.abspath('simulation_with_only_cand_of_interest/gen_1/gen_1_cand_29_cfg.json')
sim = revise_cell_params(data_path, cfg_path, temp_netParams_path, extracted_data_path, save_json=True)
print(sim.cfg)
```

```python
#Tweak SimCfg
sim.cfg.duration = 60*1000 # Duration of the simulation, in ms (60seconds)

# re-create sim object
sim.create(simConfig=sim.cfg, netParams=sim.net.params, output=True) #recreate sim object
```

```python
sim.simulate() # run parallel Neuron simulation
sim.saveData() # save results to file
```

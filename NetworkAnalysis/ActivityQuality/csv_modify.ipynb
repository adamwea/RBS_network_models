```python
import pandas as pd

# Load the uploaded CSV file
file_path = '/Users/liufanling/Library/CloudStorage/OneDrive-Personal/1 UC DAVIS/2024 Summer/CSRA/Outputs/SYNGAP_ALL/Activity/Compiled_ActivityScan.csv'
data = pd.read_csv(file_path)

# Display the first few rows to understand its structure
data.head()
```

```python
# Separate the data into two DataFrames based on the NeuronType column
# Convert Chip_ID and Well to string type and strip whitespace from NeuronType column
data['Chip_ID'] = data['Chip_ID'].astype(str)
data['Well'] = data['Well'].astype(str)
data['NeuronType'] = data['NeuronType'].str.strip()
data['Chip_ID'] = data['Chip_ID'].str.strip()
data['Well'] = data['Well'].str.strip()

wt_data = data[data['NeuronType'] == 'WT']
het_data = data[data['NeuronType'] == 'HET']
```

```python
# Save the separated data to new CSV files at the same location
wt_file_path = '/Users/liufanling/Library/CloudStorage/OneDrive-Personal/1 UC DAVIS/2024 Summer/CSRA/Outputs/SYNGAP_ALL/Activity/WT.csv'
het_file_path = '/Users/liufanling/Library/CloudStorage/OneDrive-Personal/1 UC DAVIS/2024 Summer/CSRA/Outputs/SYNGAP_ALL/Activity/HET.csv'

wt_data.to_csv(wt_file_path, index=False)
het_data.to_csv(het_file_path, index=False)
```

```python
# Define the Chip_ID and Well combinations to change NeuronType to 'none' with Well as string
combinations_to_change = [
    ('M06691', '1'), ('M06691', '4'),
    ('M07301', '1'), ('M07301', '2'), ('M07301', '4'), ('M07301', '5'),
    ('M07309', '1'), ('M07309', '2'), ('M07309', '3'),
    ('M07306', '1'), ('M07306', '3'), ('M07306', '6')
]

# Apply changes to NeuronType column based on Chip_ID and Well combinations
for chip_id, well in combinations_to_change:
    wt_data.loc[(wt_data['Chip_ID'] == chip_id) & (wt_data['Well'].astype(str) == well), 'NeuronType'] = 'none'

# Define the Chip_ID and Well combinations to change NeuronType to 'treat_a'
combinations_to_change_treat_a = [
    ('M06691', '2'), ('M06691', '3'),
    ('M07309', '5'), ('M07309', '6')
]

# Apply changes to NeuronType column based on Chip_ID and Well combinations
for chip_id, well in combinations_to_change_treat_a:
    wt_data.loc[(wt_data['Chip_ID'] == chip_id) & (wt_data['Well'].astype(str) == well), 'NeuronType'] = 'treat_a'

# Define the Chip_ID and Well combinations to change NeuronType to 'treat_b'
combinations_to_change_treat_b = [
    ('M06691', '5'), ('M06691', '6'),
    ('M07306', '2'), ('M07306', '4')
]

# Apply changes to NeuronType column based on Chip_ID and Well combinations
for chip_id, well in combinations_to_change_treat_b:
    wt_data.loc[(wt_data['Chip_ID'] == chip_id) & (wt_data['Well'].astype(str) == well), 'NeuronType'] = 'treat_b'

# Define the Chip_ID and Well combinations to change NeuronType to 'treat_c'
combinations_to_change_treat_c = [
    ('M07301', '3'), ('M07301', '6'),
    ('M07306', '5'), ('M07309', '4')
]

# Apply changes to NeuronType column based on Chip_ID and Well combinations
for chip_id, well in combinations_to_change_treat_c:
    wt_data.loc[(wt_data['Chip_ID'] == chip_id) & (wt_data['Well'].astype(str) == well), 'NeuronType'] = 'treat_c'
```

```python
# Save the modified DataFrame to a new CSV file
modified_wt = '/Users/liufanling/Library/CloudStorage/OneDrive-Personal/1 UC DAVIS/2024 Summer/CSRA/Outputs/SYNGAP_ALL/Activity/Modified_WT.csv'
wt_data.to_csv(modified_wt, index=False)
```
